USING FOR LOOP FOR REMOVE DUPLICATES

```python
def remove_duplicates(list_with_duplicates):
  unique_list = []
  for item in list_with_duplicates:
    if item not in unique_list:
      unique_list.append(item)
  return unique_list

original_list = [12,44,21,4,2,3,3,2,44]
unique_list = remove_duplicates(original_list)
print(unique_list)
```

**OUTLIER DETECTION TECHNIQUE**

```python
import numpy as np
data = [1,2,2,2,3,1,1,15,2,2,2,3,1,1,2]
mean = np.mean(data)
std = np.std(data)
print(round(mean,2))
print(round(std,2))
```

```python
thersold = 3
outliers = []
for i in data:
  z_score = (i-mean)/std
  if np.abs(z_score)>thersold:
    outliers.append(i)

print(outliers)
```
